```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("processed_data.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df["Class"].value_counts()
df["Class"].mean()*100

#Fraud transactions constitute a very small percentage indicating strong class imbalance.
```

```python
fraud_by_hour = df.groupby("Hour")["Class"].sum()
fraud_by_hour.plot(kind="bar", figsize=(10,5))
plt.title("Fraud Distribution By Hour")
plt.show()

#Fraud activity peaks during specific low-activity hours.
```

```python
df.groupby("Class")["Amount"].agg(["mean","median","std"])

#Compare mean and median.
```

```python
df[df["Class"]==1]["Amount"].hist(bins=50)
plt.title("Fraud Amount Distribution")
plt.show()
```

```python
df.groupby("Amount Category")["Class"].mean()

#High-value transactions show higher fraud probability.
```

```python
df.groupby("Time Category")["Class"].mean()

#Night transaction exhibit higher fraud rate compared to daytime transactions.
```

```python
df.groupby(["Time Category","Amount Category"])["Class"].mean()

#Fraud probability increases significantly when high-value transactions occur during night hours.
```

```python
numeric_df=df.select_dtypes(include=['number'])
corr = numeric_df.corr()["Class"].sort_values()
corr.plot(kind="barh",figsize=(8,6))
plt.title("Correlation of Features with Fraud")
plt.show()

#Certain transaction features show stronger statistical relationship with fraudulent class
```

```python
plt.boxplot(df["Amount"])
plt.title("Transaction Amount Outlier Detection")
plt.show()

#Fraudulent transactions often behave as statistical outliers.
```

```python
df["Cumulative_Fraud"]=df["Class"].cumsum()
df["Cumulative_Fraud"].plot()
plt.title("Cumulative Fraud Growth")
plt.show()

#Fraud incidents show consistent incremental growth over time
```

```python
heatmap_data=df.pivot_table(
    values="Class",
    index="Hour",
    columns="Amount Category",
    aggfunc="mean"
)
heatmap_data
```

```python
plt.figure(figsize=(8,6))
plt.imshow(heatmap_data,aspect='auto')
plt.colorbar(label="Fraud Rate")
plt.xticks(range(len(heatmap_data.columns)), heatmap_data.columns)
plt.yticks(range(len(heatmap_data.index)),heatmap_data.index)
plt.title("Fraud Rate Heatmap (Hour VS Amount Category)")
plt.xlabel("Amount Category")
plt.ylabel("Hour")
plt.show()

#fraud rate is significantly higher during late night hours for high value transactions. Low value transactions during daytime show minimal fruad probability
```

```python
fraud_ratio=df.groupby("Amount Category")["Class"].mean()*100
fraud_ratio=fraud_ratio.round(2)
fraud_ratio

#high value transactions show 2.8% fraud rate compared to 0.1% in low valur transactions
```

```python
df.groupby("Hour")["Class"].mean()*100
```

```python
# final insights gained
# 1. Fraud transactions represent a very small portion of total transactions, indicating strong class imbalance.
# 2. Fraud activity is not uniformly distributed across time, peak activity occurs during low-activity night hours.
# 3. High-value transactions exhibit significantly higher fraud rates compared to low-value transactions.
# 4. Combined behavioral risk (High Amount + Night Transaction) increases fraud probability substantially.
# 5. Fraudulent transactions show higher variability in amount compared to legitimate transactions.
# 6. Certain anonymized features (V-variables) show moderate statistical correlation with fraud class.
```
